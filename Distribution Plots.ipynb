### Distribution Plots
- Basically they are used to find the frequency of data points 
- In Dist Plot KDE Parameter is True and in Histogram Plot the Kde is False else then both are same
#### Dist Plot

```python
sns.distplot(df['spirit_servings'],bins=20)
```

```python
sns.distplot(df['spirit_servings'],kde=False,bins=20)
```

### Biavariante Plot
- That's mean we can put two parameter in plot 
#### Joint Plot

```python
sns.jointplot('spirit_servings','beer_servings',data=df,kind='hex')
```

#### Pair Plot 

```python
sns.pairplot(data=df,kind='kde')
```

```python
sns.lmplot('spirit_servings','beer_servings',data=df,hue='continent')
```
