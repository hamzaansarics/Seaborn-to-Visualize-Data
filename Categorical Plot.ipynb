### Catigorical Scatter Plot
- They are use to find relationship between two variables as catigorical and numerical variables 
###### Strip Plot

```python
sns.stripplot('Group','Number_pregnant',data=df)
```

#### Swarm Plot
- They are similar to Stripplot but only on difference is that they are not merged/overlap the data points 

```python
sns.swarmplot('Group','Number_pregnant',data=df)
```

### Catigorical Distribution Plot
- They are also used to find the relationship between catigorcal and numerical varibales but they used another graphs

#### Boxplot

```python
sns.boxplot('Age','Group',data=df)
```

```python
df['Age']=df['Age'][df.Group=='B']
```

#### Violinplot 

```python
sns.violinplot('Group','Age',data=df)
```

### Catigorical Estimation Plot
#### Point Plot
- Actually they will draw a horizental line between the vertical lines of data that's shows the mean of data points 

```python
sns.pointplot('Group','Age',data=df)
df['Age'][df.Group=='B'].max()
```

```python
sns.barplot('Group','Age',data=df)
```
