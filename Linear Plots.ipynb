```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'I:\Text Datasets\CSV\diabetes.csv')
iris=pd.read_csv('I:\Text Datasets\CSV\iris.csv')
iris.head()
```

#### Regression Plot
- They will Find Relationship between two variables with regression line

```python
sns.regplot("BMI",'Triceps',data=df,label='Reg')
plt.legend()
```

#### Linear Model Plot
 - They will also find relationship between two variables but they take extra attribute then regression plot , actually they will take third attribute exmaple gender [male / female] how much give more tip than other as male or female in tip example

```python
sns.lmplot("Age",'Blood_pressure',data=df,hue='Class')
plt.legend()
```

#### Joint Plot 
- They also tell about relationship between two variables with regrssion line and also give us distributions(how  data is distributed) for two variables

```python
sns.jointplot("Age",'Blood_pressure',data=df,kind='reg')
```

#### Pair Plot 
- They take dataframe instead of attributes and they plot all of numeric and float attributes with scatter plot and histogram 

```python
sns.pairplot(df[['Number_pregnant','Age','Triceps']])
```

#### PairGrid
- PairGrid is similar to pairplot but only difference is that we can map plots from matplotlib or seaborn as change default plots into our choice graph

```python
x=sns.PairGrid(iris,vars=['PetalLengthCm','PetalWidthCm','Species'],hue='Species')
# x.map(plt.scatter)
# x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.bar).add_legend()
x.map_diag(sns.swarmplot)
```

#### Heatmap 
 - They are most imortant in machine learing for building model , becuase they also get dataframe instead of attributes and show the strength of relatoinship between numeric and float attributes


```python
sns.heatmap(df[['Number_pregnant','Age','Triceps']].corr())
```

```python
sns.heatmap(data=df[['Blood_pressure','Age']].corr(),vmin=0,vmax=1,annot=True)
```
